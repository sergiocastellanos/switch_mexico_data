```python
import pandas as pd
import plotly.plotly as py
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, iplot
from plotly.graph_objs import *
init_notebook_mode()
import numpy as np
```

```python
df = pd.read_csv("C:/Users/Aldo/Documents/Facultad/Internship Berkeley/Datos SWITCH/hora_CSV/2016.csv",index_col=range(4),skiprows=range(4),header=0)
for i in range(2017,2031):
    a = pd.read_csv("C:/Users/Aldo/Documents/Facultad/Internship Berkeley/Datos SWITCH/hora_CSV/{0}.csv".format(i),index_col=range(4),skiprows=range(4), header = 0)
    df = df.append(a)
df = df.astype('float64')
```

```python
for k in df.columns.tolist():
    yy=[]
    zz=[]
    for d in range(1,32):
        e=df.xs(d,level=2)[k]
        for a in range(2016,2030):
            for m in range(1,13):
                try: zz.append(e.xs([a,m]).mean())
                except KeyError: pass
                
        yy.append(np.mean(zz))
    data1.append(go.Scatter(x=range(1,32),y=yy,mode="markers",marker=dict(size=22,color = "rgba({0}, {1}, {2},1)".format(np.random.randint(1, 255),np.random.randint(1, 255),np.random.randint(1, 255)),line=dict(width=1))))  
    
    layout1=dict(title='Variacin mensual escenario medio en nodo {0}'.format(k),xaxis=dict(title='Dia del mes'),yaxis=dict(title='Demanda')) 
    fig1=dict(data=data1, layout=layout1)
    iplot(fig1)
```
