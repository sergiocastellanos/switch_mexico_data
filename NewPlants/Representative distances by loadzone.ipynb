```python
import pandas as pd
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
mexico = pd.read_csv("mexico_w_lz_ratio_2.csv")
```

```python
del mexico['X']
del mexico['Y']
del mexico['field_1']
```

```python
mexico = mexico.sort_values(['lz', 'RATIO'], ascending = [True, False])
mexico
```

```python
plots = []
new_plot = []
c = -1

for i, row in mexico.iterrows():
    if (int(row['lz']) != c):
        if c != -1:
            plots.append(new_plot)
        new_plot = []
        c = int(row['lz'])
    new_plot.append(row['RATIO'])
```

```python
for p in plots:
    p = sorted(p)
    fp = int(len(p)*0.8)
    plt.plot(p[fp:])
```

```python
media = []
for p in plots:
    p = sorted(p)
    f = p[int(len(p)*0.8):]
    media.append(f[int(len(f)/2)])
```

```python
representative_distances = []
for m in media:
    for i, row in mexico.iterrows():
        if row['RATIO'] == m:
            representative_distances.append(row['NEAREST_SU'])
            break
```
